# Cumulative distribution functions

Some limitations of PMFs:

* As the number of values in a distribution -> inifinity, the effect of **random noise** increases
* Hard to **compare distributions**
* Sussing out **meaningful patterns** can be difficult
* **Binning** is a useful way of cutting noise, but can also remove significant patterns

An alternative: **cumulative distribution functions (CDFs)**!

### Definition

Mathematically, a CDF is the integral of a PDF. Downey provides an interesting alternate definition: a function that **maps values to percentiles**.

This makes sense if you think about the intuition of an integral: as the area under a curve. The CDF for any given value returns the area under the probability curve from the lower limit of the distribution to the current value – that is, the *percent of the distribution that falls in that range.*

From this definition, we can reason that a CDF will always be **monotically increasing** and have a **limit of 1**.

### Reading a CDF

The obvious way to read a CDF is to interpret it as a literal function: look up distribution values, and find their percentile.

But Downey points out that we can also use them to infer properties of the distribution. Take the CDF of pregnancy length:

```python
%matplotlib inline

import first
import nsfg
import thinkstats2
import thinkplot

live, firsts, others = first.MakeFrames()
cdf = thinkstats2.Cdf(live.prglngth)

thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='weeks', ylabel='CDF')
```

We can easily see the mode of the distribution, 39, represented as a steep vertical line. We can also see the extremely low-probability tails of the distribution: pregnancy lengths shorter than 30 weeks and longer than 45 weeks.

Let's confirm these conclusions by plotting the PMF:

```python
pmf = thinkstats2.Pmf(live.prglngth)

thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='weeks', ylabel='PMF')
```

# Exercises

*The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.*

*Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?*

```python
import random

random_dist = [random.random() for i in range(1000)]

random_pmf = thinkstats2.Pmf(random_dist)
thinkplot.Pmf(random_pmf)
thinkplot.Show(xlabel='random numbers', ylabel='PMF')

random_cdf = thinkstats2.Cdf(random_dist)
thinkplot.Cdf(random_cdf)
thinkplot.Show(xlabel='random numbers', ylabel='CDF')
```

Yup – looks uniform to me!
